## 1. Seeing the world in 3D perspective

### 1.1 Why is 3D important?

> - AI agents operate in the real world, which is a 3D space

> - AR/VR
> - 3D printing
> - Medical applications

### 1.2 The way we observe 3D

> - An image is a projection of the 3D world onto a 2D space

> - Triangulation 
    - The way to obtain a 3D point from 2D images
    
### 1.3 3D data representation

> - A 2D image is represented by RGB values of each pixel in 2D array structure
> - 3D data representation is not unique
    - Multi-view images
    - Volumetric (voxel)
    - Part assembly
    - Point cloud
    - Mesh (Graph CNN)
    - Implicit shape
    
### 1.4 3D datasets

> - ShapeNet
    - 55 categories의 51,300의 data
> - PartNet
    - Fine-grained dataset
    - 26,271개의 data 및 573,585의 instance를 가지고 있다.
> - SceneNet
    - 5million RGB-Depth synthetic indoor images
> - ScanNet
    - RGB-Depth dataset with 2.5 million
> - Outdoor 3D scene datasets (typically for autonomous vehicle applications)
    - KITTI
    - Semantic KITTi
    - Waymo Open Dataset
