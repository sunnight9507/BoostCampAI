### 1.1 Why is visual perception important?

### 1.2 What is computer vision?

> - 사람이 장면을 이해하는 과정
    - 일상의 장면을 관찰
    - 장면의 상이 맺힘
    - 이를 뇌에 전달하고 해석
    
> - 이미지 정보를 바탕으로 2D 이미지를 drawing 하는 과정을 **Computer Graphics(Rendering)**이라고 한다.
> - Computer Graphics의 반대되는 일을 **Computer Vision(Inverse rendering)**이라고 한다.

> - 사람의 시각능력도 불안전하다.
    - 어느 정도 bias 되어 학습되어 있다고 볼 수 있다.

> - Machine Learning
    - input -> Feature extraction -> Classification -> Output
> - Deep Learning
    - input -> Feature extraction + Classification -> Output

## 3. CNN architectures for image classification 1

### 3.1 Brief History

> - Alexnet -> VGGNet -> GoogLeNet -> ResNet -> Beyond ResNet

### 3.2 AlexNet

> - Yann LeCun이 개발한 간단한 CNN LeNet에서 발전
> - layer가 deep 해졌고, ImageNet을 사용
> - ReLU, dropout 사용
> - Local Response Normalization (LRN)을 사용
    - 현재는 Batch normalization을 사용
> - 첫번째 filter size를 11x11 사용

### 3.3 VGGNet

> - Deeper architecture
    - 16 and 19 layers
> - Simpler architecture
    - No NRN
    - 3x3 filtering, 2x2 max pooling만 사용
> - AlexNet보다 좋은 성능과 높은 일반화
