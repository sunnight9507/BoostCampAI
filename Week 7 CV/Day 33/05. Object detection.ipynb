## 1. Object detection

### 1.1 What is object detection?

> - Fundamental image recognition tasts
    - Semantic segmentation
    - Instance segmentation
    - Panoptic segmentation
    
> - Classification + Box localization
    - ex) (P class, X min, Y min, X max, Y max)
    
### 1.2 What are the applications of object detection?

> - Autonomous driving
> - Optical Character Recognition (OCR)

## 2. Two-stage detector

### 2.0 Traditional methods - hand-crafted techniques

> - Gradient-based detector (e.g., HOG)

> - Selective search
    - 1) Over-segmentation : 영상을 비슷한 색끼리 잘게 분할
    - 2) 비슷한 영역끼리 병합 (ex. 색, gradient, ...)
    - 3) 반복 후 큰 segmentation의 bounding box를 추출
    
### 2.1 R-CNN

> - Directly leverage image classification networks for object detection

> - Process
    - 1) Input image
    - 2) Extract region proposals (~2k) & warped region
    - 3) Compute CNN features
    - 4) Classify regions
    
> - 단점
    - 각각의 영역마다 분류를 진행하므로 속도가 느리다.
    - hand design된 region proposal을 사용하여 성능 향상에 한계가 있다.
    
### 2.2 Fast R-CNN

> - Recycle a pre-computed feature for multiple object detection

> - 1) Conv. feature map from the original image
> - 2) RoI feature extraction from the feature map through RoI pooling
> - 3) Class and box prediction for each RoI

### 2.3 Faster R-CNN

> - End-to-end object detection by neural region proposal

> - IoU (Intersection over Union)
    - $\frac{\mathrm{Area \ of \ Overlap}}{\mathrm{Area \ of \ Union}}$
    
> - Anchor boxes
    - A set of pre-defined bounding boxes
    - IoU 값에 따라 object 판단
    
> - Region Proposal Network (RPN)

> - Non-Maximum Suppression (NMS)
    - 1) Select the box with highest objectiveness score
    - 2) Compare IoU of this box with other boxes
    - 3) Remove the bounding boxes with IoU >= 50%
    - 4) Move to the next highest objectiveness score
    - 5) Repeat steps 2-4
    
## 3. Single-stage detector

### 3.0 Comparision with two-stage detectors

> - No explicit RoI pooling
> - 정확도는 조금 포기하고 간단하고 빠른 수행 시간을 보여준다.

### 3.1 You only look once (YOLO)

> - Process
    - 1) S x S grid on input
    - 2-1) Bounding boxes + confidence
    - 2-2) Class probability map
    - 3) Final detections
    
### 3.2 Single Shot MultiBox Detector (SSD)

> - 각 feature map마다 다양한 bounding box의 shape들을 고려할 수 있다.

> - Anchor box의 수는 많지만 네트워크가 단순해 YOLO보다 훨씬 빠른 속도, 높은 성능을 보여준다.

## 4. Two-stage detector vs. one-stage detector

### 4.1 Focal loss

> - Class imbalance problem
    - many negative examples, no useful signal
    - Few positive examples, rich information
    
> - Focal loss
    - cross entropy loss의 확장 version
    - $\mathrm{FL}(p_t) = - (1 - p_t)^\gamma \log{(p_t)}$
    
### 4.2 RetinaNet

> - Feature Pyramid Networks (FPN) + class/box prediction branches
