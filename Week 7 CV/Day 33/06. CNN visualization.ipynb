## 2. Analysis of model behaviors

### 2.1 Embedding feature analysis

> - DB 내의 image data set과 비교하여 유사한 image를 찾아줌
    - Nearest neighbors (NN)

> - Dimensionality reduction
    - t-distributed stochastic neighbor embedding (t-SNE)
    
### 2.2 Activation investigation

> - Layer activation을 분석
    - 각 layer의 hidden layer를 확인 할 수 있다.
    
> - Maximally activating patches
    - 1) Pick a channel in a certain layer
    - 2) image를 network에 넣어주고 각 activation 값을 기록한다.
    - 3) activation 값이 큰 영역을 원본 image에서 crop
    
> - Class visualization
    - Gradient ascent
    - Generate a synthetic image that triggers maximal class activation
    - $ I^* = \mathrm{argmax} f(I) - \lambda ||I||_2^2$
> - Process
    - 1) 임의의 image를 넣어 class score를 얻는다.
    - 2) 역전파를 통해 입력단의 gradient를 구한다. 입력 gradient가 어떻게 변해야 target class score를 높일 수 있는지 update
    - 3) update the current image
    - 1-3의 과정을 반복 진행
    
## 3. Model decision explanation

### 3.1 Saliency test

> - Occlusion map
    - image가 주어졌을 때 제대로 판정되기 위해 각 영역의 중요도를 추출하는 방법
    - Prediction scores change according to the location of mask
    
> - via Backpropagation
    - 1) 입력 image를 넣어 class score를 얻는다.
    - 2) 역전파를 input domain까지 적용
    - 3) 얻어진 값의 절대적인 크기를 visualizing
    
### 3.2 Backpropagation-based saliency

> - Rectified unit (backward pass)
    - backward pass를 진행할 때 gradient값에 relu를 적용하는 방법
    
> - Guided backpropagation
    - forward, backward의 값들에 모두 relu를 적용
    
### 3.3 Class activation mapping

> - Class activation mapping (CAM)
    - Visualize which part of image contributes to the final decision
    - Global average pooling (GAP) layer instead of the FC layer
    - By visualizing CAM, we can interpret why the network classified the input to that class
    - 단점으로는 GAP과 FC layer를 수정 후 재학습 해야된다.
    
> - Grad-CAM
    - network의 재학습, 수정 없이 CAM의 결과값을 얻는 방법
    - importance weights를 어떻게 구할지가 핵심
