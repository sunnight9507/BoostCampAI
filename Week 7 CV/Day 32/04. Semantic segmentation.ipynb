## 1. Semantic segmentation

### 1.1 What is sementic segmentation?

> - pixel 단위로 category를 분류하는 것
> - 동일한 class가 여러 개 있어도 하나의 색으로 분류

### 1.2 Where can sematic segmentation be applied to?

> - Medical images
> - Autonomous driving
> - Computational photography

## 2. Semantic segmentation architectures

### 2.1 Fully Convolutional Networks (FCN)

> - Semantic segmentation의 첫 end-to-end architecture이다.
> - 임의의 size의 image를 input으로 넣어도 작동 가능하다.

> - Fully connected vs Fully convolutional
    - FC : 공간정보를 고려하지 않고 하나의 vector를 다른 차원의 vector로 변환해준다.
    - FCN : 입력, 출력 map 형태로 나온다. 공간정보를 고려한다. 각 위치마다 classification을 수행한다.
    
> - 한계 : score map의 size가 매우 작다.
> - 이를 upsampling으로 해결

> - Upsampling
    - Transposed convolution
        - 중첩이 생기지 않도록 kernel size와 stride size를 조정해야 한다.
    - Upsample and convolution
        - Nearest-neighbor, Bilinear interpolation 후 convolution
        
> - Adding skip connections for enlarging the score map

### 2.2 Hypercolumns for object segmentation

> - Overall architecture
    - Very similar to FCN
    - 차이점 : Apply to each bounding box

### 2.3 U-Net

> - 특징
    - fully convolutional networks
    - Similar to skip connections in FCN

> - Overall architecture
    - Contracting Path
        - 일반적인 CNN과 동일
        - 채널수를 두배로 늘려주고, size는 pooling으로 줄여준다.
    - Expanding Path
        - 채널 size를 줄여주고 해상도를 올려준다.
        - Contracting Path의 동일한 층의 layer를 concatenating 시켜준다.

> - An even number is required for input and feature sizes

### 2.4 DeepLab

> - Conditional Random Fields (CRFs)
> - Dilated convolution
    - Atrous convolution
    - receptive field를 증가하게 하는 효과가 있다.
> - Depthwise separable convolution
    - 1) Depthwise convolution
    - 2) Pointwise convolution
    - 계산량을 줄여준다.
